```python
sinavSonucu = data[["G1","G2","G3"]]
sinavSonucu
```

```python
scaler = MinMaxScaler()
data = scaler.fit_transform(data)
data = pd.DataFrame(data)
data
# datanın hepsi scale edildi ama sınav sonuçları da scale edildiği için yanlış sonuç elde edilir.
```

```python
sinavSonucuScaler = (sinavSonucu - 0)/(20 - 0)
sinavSonucuScaler
# sınav sonuçlarını ayrı bir şekilde scale ettik, data ile birleştiricez.
```

```python
data = data.drop(data.columns[[22, 23, 24]],axis = 1)
data
```

```python
scalerData = data.join(sinavSonucuScaler)
scalerData
```

```python
sınavToplam = scalerData[["G1","G2","G3"]].sum(axis = 1)
sınavToplam
```

```python
scalerData = scalerData.drop(scalerData.columns[[22, 23, 24]],axis = 1)
scalerData
```

```python
scalerData["sınavToplam"] = sınavToplam
scalerData
```

```python
plt.hist(sınavToplam)
plt.show()
```

```python
# Veri setini oluşturma
# 0-1.3    1.3-1.8     1.8-3
#   0         1          2
# olacak şekilde gruplara ayrılma 
```

```python
first = scalerData["sınavToplam"] < 1.3
second = (scalerData["sınavToplam"] >= 1.3) & (scalerData["sınavToplam"] < 1.8)
third = scalerData["sınavToplam"] >= 1.8
```

```python
print(first)
```

```python
scalerData.loc[first, "sınavToplam"] = 0
scalerData.loc[second, "sınavToplam"] = 1
scalerData.loc[third, "sınavToplam"] = 2
```

```python
scalerData
```

```python
plt.hist(scalerData.sınavToplam)
plt.show()
```
